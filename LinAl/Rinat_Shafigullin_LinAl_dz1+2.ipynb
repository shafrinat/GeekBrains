### –õ–∏–Ω–µ–π–Ω–∞—è –∞–ª–≥–µ–±—Ä–∞
##### –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ ‚Ññ1

__1.__ –ò—Å—Å–ª–µ–¥—É–π—Ç–µ –Ω–∞ –ª–∏–Ω–µ–π–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

__–†–µ—à–µ–Ω–∏–µ__<br>
–ï—Å–ª–∏ –ø—Ä–∏—Å–º–æ—Ç—Ä–µ—Ç—å—Å—è –ø–æ–≤–Ω–∏–º–∞—Ç–µ–ª—å–Ω–µ–µ, —Ç–æ –º–æ–∂–Ω–æ –≤–µ–∫—Ç–æ—Ä $f_1(x)$ –≤—ã—Ä–∞–∑–∏—Ç—å —á–µ—Ä–µ–∑ –æ—Å—Ç–∞–ª—å–Ω—ã–µ —Ç—Ä–∏ –≤–µ–∫—Ç–æ—Ä–∞:
$$e^x=-(x-e^x)+(x+1)-1$$
–∏–ª–∏
$$f_1(x)=-f_4(x)+f_3(x)-f_2(x)$$
–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –ø–µ—Ä–≤—ã–π –≤–µ–∫—Ç–æ—Ä —è–≤–ª—è–µ—Ç—Å—è –∫–æ–º–±–∏–Ω–∞—Ü–∏–µ–π —Ç—Ä–µ—Ö –æ—Å—Ç–∞–ª—å–Ω—ã—Ö, –∞ –∑–Ω–∞—á–∏—Ç –æ–Ω–∏ –ª–∏–Ω–µ–π–Ω–æ –∑–∞–≤–∏—Å–∏–º—ã –º–µ–∂–¥—É —Å–æ–±–æ–π.

__2.__ –ò—Å—Å–ª–µ–¥—É–π—Ç–µ –Ω–∞ –ª–∏–Ω–µ–π–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

__–†–µ—à–µ–Ω–∏–µ__<br>
–ï—â–µ —Ä–∞–∑ —Å–º–æ—Ç—Ä–∏–º –≤–Ω–∏–º–∞—Ç–µ–ª—å–Ω–æ –Ω–∞ –¥–∞–Ω–Ω—ã–µ: $f_4(x)$ –º–æ–∂–Ω–æ —Ä–∞–∑–ª–æ–∂–∏—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–µ –∫–≤–∞–¥—Ä–∞—Ç–∞ —Å—É–º–º—ã –∏ –ø–æ–ª—É—á–∏—Ç—Å—è:
$$f_{4}(x)=(x+1)^{2}=x^2+2x+2$$
–í —Ç–∞–∫–æ–º —É–∂–µ –≤–∏–¥–µ –≤–µ–∫—Ç–æ—Ä $f_4(x)$ –º–æ–∂–Ω–æ –≤—ã—Ä–∞–∑–∏—Ç—å —á–µ—Ä–µ–∑ —Ç—Ä–∏ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞:
$$f_{4}(x)=x^2+2x+2=f_3(x)+2f_2(x)+f_1(x)$$
–ö–∞–∫ –∏ –≤ –ø–µ—Ä–≤–æ–º –∑–∞–¥–∞–Ω–∏–∏ –æ–¥–∏–Ω –≤–µ–∫—Ç–æ—Ä –º–æ–∂–Ω–æ –≤—ã—Ä–∞–∑–∏—Ç—å —á–µ—Ä–µ–∑ —Ç—Ä–∏ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö, –∞ –∑–Ω–∞—á–∏—Ç –æ–Ω–∏ –ª–∏–Ω–µ–π–Ω–æ –∑–∞–≤–∏—Å–∏–º—ã.<br>
__–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ.__ –í—ã—à–µ $2x$ –≤—ã—Ä–∞–∑–∏–ª –∫–∞–∫ $2f_2(x)$, –∞ –Ω–µ $f_1(x)\times f_2(x)$, –ø–æ—Ç–æ–º—É —á—Ç–æ –≤ –ª–∏–Ω–µ–π–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –º–æ–∂–Ω–æ —É–º–Ω–∂–∞—Ç—å –Ω–∞ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—É, –∞ –ø–µ—Ä–µ–º–Ω–æ–∂–∞—Ç—å –≤–µ–∫—Ç–æ—Ä–∞ –Ω–µ–ª—å–∑—è.

__3.__ –ù–∞–π–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤–µ–∫—Ç–æ—Ä–∞ $x = (2, 3, 5)\in \mathbb{R}^{3}$ –≤ –±–∞–∑–∏—Å–µ $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

__–†–µ—à–µ–Ω–∏–µ__<br>
$$x=b_2+3b_3+\frac{b_1}{2}$$
—Ç–æ –µ—Å—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤–µ–∫—Ç–æ—Ä–∞ $x$ –≤ –∑–∞–¥–∞–Ω–Ω–æ–º –±–∞–∑–∏—Å–µ –±—É–¥—É—Ç (1/2; 1; 3)

__4.__ –ù–∞–π–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤–µ–∫—Ç–æ—Ä–∞ $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

–∞) –≤ –±–∞–∑–∏—Å–µ $1$, $x$, $x^{2}$;

–±) –≤ –±–∞–∑–∏—Å–µ $x^{2}$, $x-1$, $1$.

__–†–µ—à–µ–Ω–∏–µ__<br>
–∞) –ó–∞–ø–∏—à–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –ø–æ—Ä—è–¥–∫—É –≤ –±–∞–∑–∏—Å–µ: 2, -2, 3. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤–µ–∫—Ç–æ—Ä–∞ –≤ –∑–∞–¥–∞–Ω–Ω–æ–º –±–∞–∑–∏—Å–µ –±—É–¥—É—Ç (2; -2; 3).
<br>
–±) –ü—Ä–æ–≤–µ–¥–µ–º –Ω–µ–±–æ–ª—å—à–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞:
$$3x^2-2x+2=3x^2-2(x-1)+0$$
–î–∞–ª–µ–µ –ø–æ —Ç–æ–π –∂–µ —Å—Ö–µ–º–µ: (3, -2, 0).

__5.__ –£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ, —Å—á–∏—Ç–∞–µ—Ç—Å—è –ª–∏ –ª–∏–Ω–µ–π–Ω—ã–º –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ–º:

–∞) —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å –≤—Å–µ—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ —Ç—Ä—ë—Ö–º–µ—Ä–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞, —É –∫–æ—Ç–æ—Ä—ã—Ö –ø–æ –∫—Ä–∞–π–Ω–µ–π –º–µ—Ä–µ –æ–¥–Ω–∞ –∏–∑ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Ä–∞–≤–Ω–∞ –Ω—É–ª—é;

–±) –≤—Å–µ –≤–µ–∫—Ç–æ—Ä—ã, —Å—á–∏—Ç–∞—é—â–∏–µ—Å—è –ª–∏–Ω–µ–π–Ω—ã–º–∏ –∫–æ–º–±–∏–Ω–∞—Ü–∏—è–º–∏ –¥–∞–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤  {ùë¢1,ùë¢2‚Ä¶,ùë¢ùëõ} .

__–†–µ—à–µ–Ω–∏–µ__<br>
a) –ß—Ç–æ–±—ã –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å –≤—Å–µ—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ —Ç—Ä—ë—Ö–º–µ—Ä–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞, —É –∫–æ—Ç–æ—Ä—ã—Ö –ø–æ –∫—Ä–∞–π–Ω–µ–π –º–µ—Ä–µ –æ–¥–Ω–∞ –∏–∑ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Ä–∞–≤–Ω–∞ –Ω—É–ª—é –ª–∏–Ω–µ–π–Ω—ã–º –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ–º, –ø—Ä–æ–≤–µ—Ä–∏–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –æ—Å–Ω–æ–≤–Ω—ã—Ö –æ–ø–µ—Ä–∞—Ü–∏–π –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞, –∞ –∏–º–µ–Ω–Ω–æ —Å–ª–æ–∂–µ–Ω–∏–µ –∏ —É–º–Ω–æ–∂–µ–Ω–∏–µ –Ω–∞ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—É.<br>
–í–æ–∑—å–º–µ–º —Ç—Ä–∏ –≤–µ–∫—Ç–æ—Ä–∞ —Å–æ —Å–ª–µ–¥—É—é—â–∏–º–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏:
$$(0, a, b), (c, 0, d), (0, 0, f)$$
–ü–æ–ø—Ä–æ–±—É–µ–º –∏—Ö —Å–ª–æ–∂–∏—Ç—å –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –ø–æ–ª—É—á–∏—Ç—Å—è:
$$(0, a, b)+(c, 0, d)+(0, 0, f)=(c, a, b+d+f)$$
–£ –Ω–∞—Å –ø–æ–ª—É—á–∏–ª–∏—Å—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤–µ–∫—Ç–æ—Ä–∞, —É –∫–æ—Ç–æ—Ä–æ–≥–æ –Ω–∏ –æ–¥–Ω–∞ –∏–∑ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –Ω–µ —Ä–∞–≤–Ω–∞ –Ω—É–ª—é. –ê –∑–Ω–∞—á–∏—Ç –Ω–∞–∑–≤–∞—Ç—å –¥–∞–Ω–Ω–æ–µ –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –ª–∏–Ω–µ–π–Ω—ã–º –Ω–µ–ª—å–∑—è.<br>
<br>
–±) –î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≤–æ–∑—å–º–µ–º –¥–≤–∞ –≤–µ–∫—Ç–æ—Ä–∞, —è–≤–ª—è—é—â–∏–µ—Å—è –ª–∏–Ω–µ–π–Ω—ã–º–∏ –∫–æ–º–±–∏–Ω–∞—Ü–∏—è–º–∏ –¥–∞–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤  {ùë¢1,ùë¢2‚Ä¶,ùë¢ùëõ}, –ø—Ä–æ–≤–µ—Ä–∏–º —Å–ª–æ–∂–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ –∏ —É–º–Ω–æ–∂–µ–Ω–∏–µ –Ω–∞ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—É.
$$a=\lambda_1\cdot u_1+\lambda_2\cdot u_2+...+\lambda_n\cdot u_n$$
$$b=\mu_1\cdot u_1+\mu_2\cdot u_2+...+\mu_n\cdot u_n$$
$$a+b=\lambda_1\cdot u_1+\mu_1\cdot u_1+\lambda_2\cdot u_2+\mu_2\cdot u_2+...+\lambda_n\cdot u_n+\mu_n\cdot u_n=(\lambda_1+\mu_1)\cdot u_1+(\lambda_2+\mu_2)\cdot u_2+...+(\lambda_n+\mu_n)\cdot u_n$$
–°–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∞–≤–∏–ª—É –¥–∏—Å—Ç—Ä–∏–±—É—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ —Å–ª–æ–∂–µ–Ω–∏—è —Å–∫–∞–ª—è—Ä–æ–≤. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Å–ª–æ–∂–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è–µ—Ç —É—Å–ª–æ–≤–∏—é –ª–∏–Ω–µ–π–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞.<br>
–¢–µ–ø–µ—Ä—å –ø—Ä–æ–≤–µ—Ä–∏–º —É–º–Ω–æ–∂–µ–Ω–∏–µ –Ω–∞ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—É:
$$\alpha \cdot a=\alpha \cdot \lambda_1\cdot u_1+\alpha \cdot \lambda_2\cdot u_2+...+\alpha \cdot \lambda_n\cdot u_n=(\alpha \cdot \lambda_1)u_1+(\alpha \cdot \lambda_2)u_2+...+(\alpha \cdot \lambda_n)u_n$$
–°–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∞–≤–∏–ª—É –∞—Å—Å–æ—Ü–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–ø–µ—Ä–∞—Ü–∏–∏ —É–º–Ω–æ–∂–µ–Ω–∏—è –Ω–∞ —Å–∫–∞–ª—è—Ä. –£–º–Ω–æ–∂–µ–Ω–∏–µ –Ω–∞ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—É —Ç–∞–∫–∂–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è–µ—Ç —É—Å–ª–æ–≤–∏—é –ª–∏–Ω–µ–π–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞.<br>
–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –¥–∞–Ω–Ω—ã–π –≤–∞—Ä–∏–∞–Ω—Ç —Å—á–∏—Ç–∞–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω—ã–º –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ–º.

##### –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ ‚Ññ2

__1.__ –ù–∞–π–¥–∏—Ç–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ $x, y \in \mathbb{R}$:<br>
–∞) $x=(0,-3, 6),~y=(-4, 7, 9);$<br>
–±) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$

__–†–µ—à–µ–Ω–∏–µ__<br>
a) <i>–í—Ä—É—á–Ω—É—é</i><br>
$(x,y)=0\times (-4)+(-3)\times 7 + 6\times9=-21+54=33$

```python
# —Ä–µ—à–µ–Ω–∏–µ –≤ –∫–æ–¥–µ
import numpy as np

x_a = np.array([0, -3, 6])
y_a = np.array([-4, 7, 9])

print(f'–°–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ x –∏ y: {x_a.dot(y_a)}')
```

–±) <i>–í—Ä—É—á–Ω—É—é</i><br>
$(x,y)=7\times (-3)+(-4)\times 1 + 0\times 11 + 1\times 2=-25+2=-23$

```python
# —Ä–µ—à–µ–Ω–∏–µ –≤ –∫–æ–¥–µ

x_b = np.array([7, -4, 0, 1])
y_b = np.array([-3, 1, 11, 2])

print(f'–°–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ x –∏ y: {x_b.dot(y_b)}')
```

__2.__ –ù–∞–π–¥–∏—Ç–µ –Ω–æ—Ä–º—ã –≤–µ–∫—Ç–æ—Ä–æ–≤ $(4, 2, 4)$ –∏ $(12, 3, 4)$ –∏ —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏.

__–†–µ—à–µ–Ω–∏–µ__<br>
<i>–í—Ä—É—á–Ω—É—é</i><br>
–î–ª—è –ø—Ä–æ—Å—Ç–æ—Ç—ã –æ–±–∑–æ–≤–µ–º –≤–µ–∫—Ç–æ—Ä–∞ –±—É–∫–≤–∞–º–∏:

m = (4, 2, 4)
n = (12, 3, 4)

–û–ø—Ä–µ–¥–µ–ª–∏–º –º–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫—É—é –Ω–æ—Ä–º—É –≤–µ–∫—Ç–æ—Ä–æ–≤:

$$\left\|m\right\|_{1} = |4|+|2|+|4|=10$$
$$\left\|n\right\|_{1} = |12|+|3|+|4|=19$$

–û–ø—Ä–µ–¥–µ–ª–∏–º —ç–≤–∫–ª–∏–¥–æ–≤—É –Ω–æ—Ä–º—É –≤–µ–∫—Ç–æ—Ä–æ–≤:

$$\left\|m\right\|_{2} = \sqrt{4^2+2^2+4^2}=\sqrt{36}=6$$
$$\left\|n\right\|_{2} = \sqrt{12^2+3^2+4^2}=\sqrt{169}=13$$

–û–ø—Ä–µ–¥–µ–ª–∏–º —É–≥–æ–ª –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏:

$$cos\alpha = \frac{4\cdot12+2\cdot3+4\cdot4}{6\cdot 13}=\frac{70}{78}\approx 0.897$$
$$\alpha \approx 26,23^{\circ} –∏–ª–∏ 0,46 —Ä–∞–¥–∏–∞–Ω$$

```python
# —Ä–µ—à–µ–Ω–∏–µ –≤ –∫–æ–¥–µ
from numpy.linalg import norm
from math import *

m = np.array([4, 2, 4])
n = np.array([12, 3, 4])

print(f'l1 –ú–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∞—è –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ m: {norm(m, ord=1)}')
print(f'l1 –ú–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∞—è –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ n: {norm(n, ord=1)}')

print(f'l2 –≠–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ m: {norm(m)}')
print(f'l2 –≠–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ n: {norm(n)}')

cos_alpha = m.dot(n) / norm(m) / norm(n)
print(f'–ö–æ—Å–∏–Ω—É—Å —É–≥–ª–∞ –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏ m –∏ n: {cos_alpha:.3f}')
print(f'–£–≥–æ–ª –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏ m –∏ n: {acos(cos_alpha):.2f} —Ä–∞–¥–∏–∞–Ω')
```

__3.__ –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –±—É–¥–µ—Ç –ª–∏ –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –µ–≤–∫–ª–∏–¥–æ–≤—ã–º, –µ—Å–ª–∏ –∑–∞ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –ø—Ä–∏–Ω—è—Ç—å:<br>
