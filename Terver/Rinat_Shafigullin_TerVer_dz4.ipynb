<b>Задача 1</b>

О непрерывной равномерно распределенной случайной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

Зная, что НСВ "B" распределена равномерно, ее D(x)=0.2, а левая граница (т.е. "а") равна 0,5, подставим все в формулу дисперсии и выразим правую границу, т.е. "b".
$$
D(x)=\frac{(b-a)^2}{12};$$
отсюда получим:
$$
b=\sqrt{12\cdot D(x)}+a=\sqrt{12\cdot 0.2}+0.5\approx 2.049$$

<b>Задача 2</b>

Рост взрослого населения города X имеет нормальное распределение, причем, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
1. больше 182 см?
2. больше 190 см?
3. от 166 см до 190 см?
4. от 166 см до 182 см?
5. от 158 см до 190 см?
6. не выше 150 см или не ниже 190 см?
7. не выше 150 см или не ниже 198 см?
8. ниже 166 см?

Задачу можно решить двумя способами: без использования сторонних библиотек (numpy, scipy, pandas и пр.), а затем проверить себя с помощью встроенных функций

Способ 1.<br>
Здесь мы имеем дело с нормальным распределеннием, с известными параметрами $\sigma$ и $\mu$. Воспользуемся правилом "трех сигм" и представим график функции плотности распределения (слайд №14 презентации урока).
1. Двигаясь вправо от $\mu$ на величину одной сигмы вероятность, что случайный человек имеет рост больше 182 см равна:
(100-68,3)/2=15,85%. Делим на 2 потому что нас интересует только правая сторона.
2. По тому же принципу: (100-95,4)/2=2,3%
3. 68,3+(95,4-68,3)/2=81,85%
4. 68,3%
5. 95,4%
6. (100-99,7)/2 + (100-95,4)/2 = 0,15+2,3=2,45%
7. 100-99,7=0,3%
8. (100-68,3)/2=15,85%.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats

# Сгенерируем выборку с известными параметрами
loc = 174
scale = 8
city_x = np.random.normal(loc, scale, size=100000) # большой город
```

```python
#Способ 2 расчеты вручную
q1 = city_x[city_x>182].shape[0]
print(f'Вероятность роста выше 182 см равна {(q1/city_x.shape[0])*100:.3f}%')

q2 = city_x[city_x>190].shape[0]
print(f'Вероятность роста выше 190 см равна {(q2/city_x.shape[0])*100:.3f}%')

q3 = city_x[(city_x>166)&(city_x<190)].shape[0]
print(f'Вероятность роста от 166 до 190 см равна {(q3/city_x.shape[0])*100:.3f}%')

q4 = city_x[(city_x>166)&(city_x<182)].shape[0]
print(f'Вероятность роста от 166 до 182 см равна {(q4/city_x.shape[0])*100:.3f}%')

q5 = city_x[(city_x>158)&(city_x<190)].shape[0]
print(f'Вероятность роста от 158 до 190 см равна {(q5/city_x.shape[0])*100:.3f}%')

q6 = city_x[(city_x<=150)|(city_x>=190)].shape[0]
print(f'Вероятность роста не выше 150 или не ниже 190 см равна {(q6/city_x.shape[0])*100:.3f}%')

q7 = city_x[(city_x<=150)|(city_x>=198)].shape[0]
print(f'Вероятность роста не выше 150 или не ниже 198 см равна {(q7/city_x.shape[0])*100:.3f}%')

q8 = city_x[city_x<166].shape[0]
print(f'Вероятность роста ниже 166 см равна {(q8/city_x.shape[0])*100:.3f}%')
```

```python
# Способ №3 со встроенными функциями
city_x2 = stats.norm(loc, scale)
qu1 = 1- city_x2.cdf(182)
print(f'Вероятность роста выше 182 см равна {qu1*100:.3f}%')

qu2 = 1- city_x2.cdf(190)
print(f'Вероятность роста выше 190 см равна {qu2*100:.3f}%')

qu3 = city_x2.cdf(190) - city_x2.cdf(166)
print(f'Вероятность роста от 166 до 190 см равна {qu3*100:.3f}%')

qu4 = city_x2.cdf(182) - city_x2.cdf(166)
print(f'Вероятность роста от 166 до 182 см равна {qu4*100:.3f}%')

qu5 = city_x2.cdf(190) - city_x2.cdf(158)
print(f'Вероятность роста от 158 до 190 см равна {qu5*100:.3f}%')

qu6 = city_x2.cdf(150) + 1- city_x2.cdf(190)
print(f'Вероятность роста не выше 150 или не ниже 190 см равна {qu6*100:.3f}%')

qu7 = city_x2.cdf(150) + 1- city_x2.cdf(198)
print(f'Вероятность роста не выше 150 или не ниже 198 см равна {qu7*100:.3f}%')

qu8 = city_x2.cdf(166)
print(f'Вероятность роста ниже 166 см равна {qu8*100:.3f}%')
```

<b> Задача 3 (Дополнительно)</b>

Коробки с шоколадом упаковываются автоматически. Их средняя масса равна 1.06 кг. Известно, что 5% коробок имеют массу, меньшую 1 кг. Найдите: а) среднее квадратическое отклонение массы коробки, б) процент коробок, имеющих массу больше 1.1 кг.

Подсказка. 1. Можно считать, что распределение массы коробки нормальное. 2. Найдите такое значение scale, для которого значение cdf(x=1, loc=1.06, scale=scale) близко к 0.05. Точности 0.0001 будет достаточно.

```python
# Сгенерируем выборку с известными параметрами
loc = 1.06
scale = 0.0001
choko_boxes = stats.norm(loc, scale) # коробки с шоколадом

# вопрос а)
while choko_boxes.cdf(1)<0.05:
    scale += 0.0001
    choko_boxes = stats.norm(loc, scale)
    choko_boxes.cdf(1)
print (f'Среднее квадратическое отклонение массы коробки равно: {scale:.4f} гр')
print (f'Процент коробок с массой менее 1 кг: {choko_boxes.cdf(x=1)*100:.4f}%')

# вопрос б)
big_box = 1- choko_boxes.cdf(1.1)
print (f'Процент коробок с массой более 1,1 кг: {big_box*100:.4f}%')
```
