```python
#Функции для использования
import math as m

# Размещение
def arrangements(k, n):
    a = m.factorial(n) // m.factorial(n - k)
    return a

# Сочетания
def combinations(k, n):
    c = m.factorial(n) // (m.factorial(k)*m.factorial(n - k))
    return c

# Перестановки
def permutations(n):
    p = m.factorial(n)
    return p

# Вероятность
def probability(m, n):
    return m/n
```

<b>Задача 2</b>

Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

Решение:

"Хотя бы один туз" означает что тузов может быть 1 или 2 или 3 или 4. И порядок карт не имеет значения, так как в задаче об этом не сказано. Таким образом мы имеем дело с сочетанием. По аналогии с примером из 1 урока условно разобьем колоду на 2 половинки, в одной 4 туза, в другой 48 остальных карт. Посчитаем сочетания для каждого случая (1, 2, 3, 4 туза) и просуммируем их количество ("или" значит +). Все 4 случая являются несовместными событиями, так как рассматриваются по отдельности.

```python
# 1 туз и 3 прочие карты
t1 = combinations(1, 4)
c3 = combinations(3, 48)
print(f'комбинаций с одним тузом: {t1*c3}')
# 2 туза и 2 прочие карты
t2 = combinations(2, 4)
c2 = combinations(2, 48)
print(f'комбинаций с двумя тузами: {t2*c2}')
# 3 туза и 1 прочая карта
t3 = combinations(3, 4)
c1 = combinations(1, 48)
print(f'комбинаций с тремя тузами: {t3*c1}')
# 4 туза и 0 прочиx карты
# Я понимаю, что комбинация всего одна, но пусть функция поработает, зря писал что-ли
t4 = combinations(4, 4)
c0 = combinations(0, 48)
print(f'комбинаций с четырьмя тузами: {t4*c0}')
# Ответ на задачу:
print(f'Число исходов, соответствующих тому, что был вытянут хотя бы один туз: {t1*c3+t2*c2+t3*c1+t4*c0}')
```

<b>Задача 3</b>

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что два определённых человека будут сидеть рядом?

Для начала, надо понять, чего от нас хочет задача. Вероятность, это отношение количества благоприятных исходов к общему количеству исходов. Общее количество исходов - это перестановки 7 человек, т.е. факториал 7. Количество благоприятных исходов равно 12 умножить на 5!. 
Поясняю:
пусть А и Б - определенные люди, 0 - остальные люди. Тогда их рассадка рядом возможна так:
1) АБ00000
2) 0АБ0000
3) 00АБ000
4) 000АБ00
5) 0000АБ0
6) 00000АБ
плюс еще 6 вариантов, если А и Б поменять местами. Но ведь остальные люди тоже садятся в произвольном порядке, а это перестановки, то есть 5!
Итого:
$$
P(A)=\frac{m}{n}=\frac{12\cdot5!}{7!}=\frac{12}{42}=\frac{2}{7}$$
<br>
Хм, забавный ответ, надеюсь, что правильный

<b>Задача 4</b>

Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает: а) все? б) два?

```python
# Для начала посчитаем общее количество исходов. Это сочетания 3 вопросов из 60.
all_comb = combinations(3, 60)
print (all_comb)
```

Необходимо посчитать количество благоприятных исходов. Тут очень похоже на задачу №2 с тузами. Разобъем все вопросы на 2 группы: студень знает 50 вопросов(win) и студент не знает 10 вопросов (fail). И посчитаем все комбинации:

```python
# Случай а, студент знает 3 ответа на 3 вопроса
win = combinations(3, 50)
fail = combinations(0, 10)
total_a = win*fail
print(f'Вероятность вытянуть 3 счастливых вопроса из 3 равна: {probability(total_a, all_comb)}')
```

```python
# Случай б, студент знает 2 ответа из 3 вопросов
win = combinations(2, 50)
fail = combinations(1, 10)
total_a = win*fail
print(f'Вероятность вытянуть 2 счастливых вопроса из 3 равна: {probability(total_a, all_comb)}')
```

<b>Задача 5</b>

Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

Чтобы доказать независимость событий А и В нужно доказать, что $P(A|B) = P(A)$. При этом условная вероятность рассчитывается по формуле $P(A|B) = \frac{P(AB)}{P(B)}$<br>
Начнем рассуждать. Вероятность события А: Р(А)=1/2, так как четных чисел на игральной кости ровно половина. Вероятность события В: Р(В)=1/2, так как чисел больше 3 на кости всего 3 (это 4, 5, 6). Вероятность одновременного происхождения этих событий: P(AB)=1/3, так как это только 2 числа из 6 (4 и 6).<br>
Посчитаем условную вероятность:
$$
P(A|B) = \frac{P(AB)}{P(B)}=\frac{\frac{1}{3}}{\frac{1}{2}}=\frac{2}{3}$$
$P(A|B)\neq P(A)$, а значит события зависимые.

<b>Задача 6 (*)</b>

Допустим, имеется некоторая очень редкая болезнь (поражает 0.1 % населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99 % больных этой болезнью и всего лишь в 1 % случаев даёт ложный положительный ответ.

Вопрос: какова вероятность, что вы действительно больны ей?<br>
Подсказка: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

Обозначим событие А - заболевание редкой болезнью, Р(А)=0,001. Событие В - тест положительный. Вероятность события В посчитаем по формуле полной вероятности:
$$
P(B)=P(B|H_1)\cdot P(H_1)+P(B|H_2)\cdot P(H_2),$$<br> где $H_1 -$ тест верный, $H_2 -$ тест ложный. $P(H_1)=0,99$, $P(H_1)=0,01$. <br>
<br>
$P(B|H_1)=0,5$, так как общее число исходов 2 (положительный или отрицательный), $P(B|H_2)=1$, так как при ложном тесте результат всегда положительный.
Тогда, подставив эти числа в формулу полной вероятности получим:
$$
P(B)=P(B|H_1)\cdot P(H_1)+P(B|H_2)\cdot P(H_2)=0,5\cdot 0,99+1\cdot 0,01=0,495+0,01=0,505$$
<br>
Найдем условную вероятность события А по формуле Байеса:<br>
$$
P(A|B)=\frac{P(B|A)\cdot P(A)}{P(B)}$$
При этом $P(B|A)=1$, то есть если человек болен (событие А), тест будет всегда положительным (событие В), не важно верный он или ложный. Осталось посчитать:
$$
P(A|B)=\frac{1\cdot 0,001}{0,505}=0,00198$$ или 0,198%
