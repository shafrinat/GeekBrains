1. Исследовать функцию на условный экстремум
$$U=3-8x+6y, \quad if \; x^2+y^2=36 $$
Составим функцию Лагранжа:
$$L(\lambda, x, y)=3-8x+6y+\lambda(x^2+y^2-36)$$
Исследуем функцию Лагранжа на экстремум, для этого начнем с первых частных производных:
Необходимое условие экстремума - производные равны 0:
$$\begin{cases}
L'_x=-8+2\lambda x=0,\\
L'_y=6+2\lambda y=0,\\
L'_{\lambda}=x^2+y^2-36=0;
\end{cases}
\begin{cases}
x=\frac{4}{\lambda},\\
y=-\frac{3}{\lambda},\\
x^2+y^2-36=0;
\end{cases}
$$
Подставляем x и y, выраженные через лямбда и решаем полученное уравнение:
$$\left(\frac{4}{\lambda}\right)^2+\left(-\frac{3}{\lambda}\right)^2=36$$
$$\frac{16}{\lambda^2}+\frac{9}{\lambda^2}=36$$
$$\frac{25}{\lambda^2}=36$$
$$\lambda^2=\frac{25}{36}$$
$$\lambda=\pm\frac{5}{6}$$
Найдем х и у:
$$x_1=4:\frac{5}{6}=\frac{24}{5}$$
$$y_1=-3:\frac{5}{6}=-\frac{18}{5}$$
$$x_2=4:-\left(\frac{5}{6}\right)=-\frac{24}{5}$$
$$y_2=-3:-\left(\frac{5}{6}\right)=\frac{18}{5}
$$
У нас получилось 2 точки с координатами:
$$
T_1\left(\frac{5}{6}, \frac{24}{5}, -\frac{18}{5}\right); T_2\left(-\frac{5}{6}, -\frac{24}{5}, \frac{18}{5}\right)
$$
Далее переходим к достаточному условию эктремума, для этого составим матрицу из вторых производных.
$$L''_{xx}=(-8+2\lambda x)'_x=2\lambda$$
$$L''_{yy}=(6+2\lambda y)'_y=2\lambda$$
$$L''_{xy}=(-8+2\lambda x)'_y=0$$
$$L''_{\lambda \lambda}=(x^2+y^2-36)'_{\lambda}=0$$
$$L''_{\lambda x}=(x^2+y^2-36)'_x=2x$$
$$L''_{\lambda y}=(x^2+y^2-36)'_y=2y
$$
$$
A=\begin{pmatrix}
0 & 2x & 2y\\
2x & 2\lambda & 0\\
2y & 0 & 2\lambda 
\end{pmatrix}
$$
Найдем определитель матрицы А:
$$\Delta_A=0\cdot 2\lambda\cdot 2\lambda+2x\cdot0\cdot2y+2x\cdot0\cdot2y-2y\cdot2\lambda\cdot2y-0\cdot0\cdot0-2x\cdot2x\cdot2\lambda=-8\lambda x^2-8\lambda y^2$$
Теперь подставим значения точек Т1 и Т2 и посчитаем значение определителя матрицы для каждой точки.<br>
Для точки Т1:
$$
\Delta_{A(Т1)}=-8\cdot\frac{5}{6}\cdot\left(\frac{24}{5}\right)^2-8\cdot\frac{5}{6}\cdot\left(-\frac{18}{5}\right)^2=-153,6-86,4=-240<0
$$
Для точки Т1 определитель получился меньше 0, значит точка T1 является условным минимумом.<br>
Для точки Т2:
$$
\Delta_{A(Т2)}=-8\cdot\left(-\frac{5}{6}\right)\cdot\left(-\frac{24}{5}\right)^2-8\cdot\left(-\frac{5}{6}\right)\cdot\left(\frac{18}{5}\right)^2=153,6+86,4=240>0
$$
Для точки Т2 определитель получился меньше 0, значит точка T2 является условным максимумом.<br>

2. Исследовать функцию на условный экстремум
$$
U=2x^2+12xy+32y^2+15\quad if \; x^2+16y^2=64
$$
Составим функцию Лагранжа:
$$
L(\lambda, x, y)=2x^2+12xy+32y^2+15+\lambda(x^2+16y^2-64)
$$
Исследуем функцию Лагранжа на экстремум, для этого начнем с первых частных производных:
Необходимое условие экстремума - производные равны 0:
$$
\begin{cases}
L'_x=4x+12y+2\lambda x=0,\\
L'_y=12x+64y+32\lambda y=0,\\
L'_{\lambda}=x^2+16y^2-64=0;
\end{cases}
$$
Выразим из первого уравнения лямбда через х и у, подставим во второе и останется система из двух уравнений с двумя неизвестными.
$$
\lambda=\frac{-4x-12y}{2x}=\frac{-2x-6y}{x}$$
$$12x+64y+32\lambda y=0,$$
$$3x+16y+8\lambda y=0,$$
$$3x+16y+8y\left(\frac{-2x-6y}{x}\right)=0,$$
$$3x+16y-16y-\frac{48y^2}{x}=0,$$
$$3x^2-48y^2=0,$$
$$x^2=16y^2.$$
Полученная система из 2-х уравнений имеет вид:
$$
\begin{cases}
x^2=16y^2,\\
x^2+16y^2-64=0;
\end{cases}
$$
Подставим х2 из первого уравнения во второе и получим:
$$
32y^2=64,$$
$$y^2=2,$$
$$y=\pm\sqrt{2}
$$
Найдем x:
$$
x^2=16\cdot2,$$
$$x=\pm4\sqrt{2}
$$
Теперь найдем все лямбды (аж 4 штуки):
$$
\lambda_1=\frac{-2\cdot4\sqrt{2}-6\cdot\sqrt{2}}{4\sqrt{2}}=-\frac{7}{2};$$
$$\lambda_2=\frac{-2\cdot4\sqrt{2}-6\cdot(-\sqrt{2})}{4\sqrt{2}}=-\frac{1}{2}$$
$$\lambda_3=\frac{-2\cdot(-4\sqrt{2})-6\cdot\sqrt{2}}{(-4\sqrt{2})}=-\frac{1}{2}$$
$$\lambda_4=\frac{-2\cdot(-4\sqrt{2})-6\cdot(-\sqrt{2})}{(-4\sqrt{2})}=-\frac{7}{2}.
$$
В результате после решения системы уравнений мы получили 4 стационарные точки со следующими координатами:
$$
T_1\left(-\frac{7}{2}, 4\sqrt{2}, \sqrt{2}\right); \quad T_2\left(-\frac{1}{2}, 4\sqrt{2}, -\sqrt{2}\right)$$
$$T_3\left(-\frac{1}{2}, -4\sqrt{2}, \sqrt{2}\right); T_4\left(-\frac{7}{2}, -4\sqrt{2}, -\sqrt{2}\right)
$$
Далее переходим к достаточному условию эктремума, для этого составим матрицу из вторых производных.
$$
L''_{xx}=(4x+12y+2\lambda x)'_x=4+2\lambda;$$
$$L''_{yy}=(12x+64y+32\lambda y)'_y=64+32\lambda;$$
$$L''_{xy}=(4x+12y+2\lambda x)'_y=12;$$
$$L''_{\lambda\lambda}=(x^2+16y^2-64)'_{\lambda}=0;$$
$$L''_{\lambda x}=(x^2+16y^2-64)'_x=2x;$$
$$L''_{\lambda y}=(x^2+16y^2-64)'_y=32y.
$$
$$
A=\begin{pmatrix}
0 & 2x & 32y\\
2x & 4+2\lambda & 12\\
32y & 12 & 64+32\lambda 
\end{pmatrix}
$$
найдем определитель:
$$
\Delta_A=0\cdot(4+2\lambda)\cdot(64+32\lambda)+2x\cdot12\cdot32y+2x\cdot12\cdot32y-32y\cdot(4+2\lambda)\cdot32y-12\cdot12\cdot0-2x\cdot2x\cdot(64+32\lambda)=\\
=1536xy-4096y^2-2048\lambda y^2-256x^2-128\lambda x^2=1536xy-2048y^2(2+\lambda)-128x^2(2+\lambda)=\\
=1536xy-128\cdot(2+\lambda)(16y^2+x^2)=128(12xy-(2+\lambda)(16y^2+x^2))
$$

```python
#Дальше в дело вступает Python, зря мы его учили чтоли...
import math as m

def delta(t):
    delta_a=128*(12*t[1]*t[2]-(2+t[0])*(16*t[2]*t[2]+t[1]*t[1]))
    return delta_a

t1=[-3.5, 4*m.sqrt(2), m.sqrt(2)]
t2=[-0.5, 4*m.sqrt(2), -(m.sqrt(2))]
t3=[-0.5, -(4*m.sqrt(2)), m.sqrt(2)]
t4=[-3.5, -(4*m.sqrt(2)), -(m.sqrt(2))]
t_dots=[t1, t2, t3, t4]

for i in range(len(t_dots)):
    print(f'Определитель матрицы для точки T{i+1} равен {delta(t_dots[i]):.2f}')
```

Как мы видим из результатов вышепроведенных вычислений, точки Т1 и Т4 являются условными максимумами, а точки Т2 и Т3 являются условными минимумами.

U'_c=16\cdot\left(-\frac{9}{17}\right)-24\cdot\frac{8}{17}+18\cdot\left(-\frac{12}{17}\right)=-\frac{552}{17}
