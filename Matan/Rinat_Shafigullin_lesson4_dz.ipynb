### Тема: “Предел функции”

##### 1 Предложить пример функции, не имеющей предела в нуле и в бесконечностях.

$$f(x)=\frac {sinx}{x}$$

x в знаменателе, таким образом в точке "0" функция не имеет предела. 
Функция синуса периодическая, поэтому не имеет пределов в бесконечностях.

##### 2 Привести пример функции, не имеющей предела в точке, но определенной в ней.

Например:

$$f(x) = \begin{cases}
                x^2-1, & \quad \text{при } x\geq 0, \\
                -x^2+1, & \quad \text{при } x<0 
            \end{cases}$$<br>
            
Отсутствие предела функции в точке означает, что для любого заданного значения области значений можно подобрать такую окрестность этого значения, что в любой сколь угодно малой окрестности точки, в которой функция принимает заданное значение, существуют точки, значение функции в которых окажется за пределами указанной окрестности. Такой точкой для указанной функции является x=0, функция не имеет левостороннего предела при x→-0.<br>
Также, примером такой функции является функция Дирихле:
$$D(x)=\begin{cases}
                1, & \quad x \in {Q}, \\
                0, & \quad x \in {R}/{Q}, 
            \end{cases}$$<br>
которая принимая значения 1 при рациональных значениях <i>x</i>, и значения 0, при иррацональных значения <i>x</i>.

##### Задание №3
Исследовать функцию
$$f(x)=x^3-x^2  $$<br>
по плану:<br>
a. Область задания и область значений.<br>
b. Нули функции и их кратность.<br>
c. Отрезки знакопостоянства.<br>
d. Интервалы монотонности* (можно не делать, требует графика или знания производной)<br>
e. Четность функции.<br>
f. Ограниченность.<br>
g. Периодичность.<br>
<br>
a. Так как в функции отсутствует знаменатель, и не наложены какие-либо иные ограничения на x, то областью задания функции является $$(-\infty;+\infty)$$<br>
<br>
b. Если решить уравнение:
$$x^3-x^2=0,$$
то получим корни x1=0, x2=1. Чтобы выяснить кратность корней, разложим многочлен на множители:
$$x^3-x^2=x\times x\times(x-1)$$
Таким образом x1=0 - корень кратности 2, х2=1 - корень кратности 1.
<br>
c. Отрезки знакопостоянства. Пойду простым путем - использую метод подбора. Как решать неравенства по школьной программе не помню. Так как известны корни уравнения и это x1=0, x2=1, это точки, в которых график функции пересекает ось Ох. Переберем по одной точке из промежутка x<0, 0<x<1, x>1. Получим:
$$\begin{cases}
                x^3-x^2=-2, \quad \text{for} \: x=-1, \\
                x^3-x^2=-0.125, \quad \text{for} \: x=0.5, \\
                x^3-x^2=4, \quad \text{for} \: x=2, \\
            \end{cases}$$<br>
Таким образом, f(x)<0 в интервалах: $$(-\infty;0) \cup (0;1),$$<br>
и f(x)>0 на интервале: $$(1;+\infty),$$<br>
<br>
d. А производные не забыл.<br>
$$f'(x)=3x^2-2x$$<br>
Производная имеет вид квадратичной функции, графиком является парабола. Множитель а>0, значит график параболы смотрит "рогами вверх", свободный член с=0, значит парабола не смещена по оси Oy. Корни х=0 и х=-2/3, значит парабола пересекает ось Ох в этих точка и производная принимает отрицательные значения между ними. Таким образом, значения производной положительны на промежутках:
$$(-\infty;0) \cup (\frac{2}{3};+\infty),$$<br>
значит исходная функция на этих промежутках монотонно возрастает.<br>
Значения производной отрицательны на промежутке:$$(0;\frac{2}{3}),$$
значит исходная функция на этом промежутке монотонно убывает.
<br>
e. Для проверки функции на четность/нечетность продставим -x вместо х:
$$f(-x)=(-x)^3-(-x)^2=-x^3-x^2; f(-x)\neq f(x)$$
Функция не вернулась к исходному виду, значит она не является четной. Попробуем вынести минус за скобки и проверить ее на нечетность:
$$f(-x)=(-x)^3-(-x)^2=-x^3-x^2=-(x^3+x^2); f(-x)\neq -f(x)$$
Выражение в скобках не равняется исходной функции, значит функция не является нечетной. Таким образом функция f(x) является функцией общего вида.<br>
f. Исследования на ограниченность.<br>
$$ lim_{x \to +\infty}x^3-x^2=\infty - \infty$$
Получили неопределенность, хм... Вынесем x^2 за скобки и получим:
$$lim_{x \to +\infty}x^2(x-1)=\infty \times \infty=\infty$$
Результат - справа функция не ограничена.
$$ lim_{x \to -\infty}x^3-x^2=-\infty - \infty$$
Снова неопределенность.
$$lim_{x \to -\infty}x^2(x-1)=\infty \times (-\infty)=-\infty$$
Результат - слева функция тоже не ограничена.<br>
g. Исследование на периодичность.<br>
Допустим, что Т-период функции, тогда для любого x справедливо соотношение:
$$x^3-x^2=(x+T)^3-(x+T)^2$$
Если x=0, то:
$$T^3-T^2=0$$
Если x=-T, то:
$$(-T)^3-(-T)^2=0 \\
T^3+T^2=0$$
Объединим оба уравнения в систему и найдем Т:
$$\begin{cases}
T^3-T^2=0,\\
T^3+T^2=0;
\end{cases}$$
Сложим оба уравнения, чтобы избавиться от Т в квадрате:
$$2T^3=0,\\
T=0$$
Таким образом, соотношение 
$$x^3-x^2=(x+T)^3-(x+T)^2$$ справедливо при T=0. Отсюда следует, что функция не периодичная.

##### Задание №4
Найти предел:<br>
a. $$lim_{x \to 0}\frac{3x^3-2x^2}{4x^2}$$
b. $$lim_{x \to 0}\frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1}$$
<br>
Поехали:
a. $$lim_{x \to 0}\frac{3x^3-2x^2}{4x^2}=lim_{x \to 0}\frac{x^2(3x-2)}{4x^2}=lim_{x \to 0}\frac{(3x-2)}{4}=-\frac{2}{4}=--\frac{1}{2}$$
<br>
b. $$lim_{x \to 0}\frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1}$$
Домножим и разделим на сопряженное выражение числитель, чтобы собрать в разность квадратов, а также домножим и разделим на сопряженное выражение знаменатель, чтобы собрать разность кубов:
$$lim_{x \to 0}\frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1}=lim_{x \to 0}\frac{(\sqrt{1+x}-1)(\sqrt{1+x}+1)\times(\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1)}{(\sqrt{1+x}+1)\times(\sqrt[3]{1+x}-1)(\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1)}=lim_{x \to 0}\frac{(1+x-1)\times(\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1)}{(\sqrt{1+x}+1)\times(1+x-1)}=lim_{x \to 0}\frac{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}{\sqrt{1+x}+1}=\frac{1+1+1}{1+1}=\frac{3}{2}$$
