##### 1. Как относятся друг к другу множество и последовательность? (в ответе использовать слова типа: часть, целое, общее, частное, родитель, дочернний субъект и т.д.)

На мой взгляд - последовательность есть набор элементов некоторого множества и по отношению к нему является дочерним субъектом. Например последовательность нечетных чисел {1, 3, 5, 7 ... n, n+2} является частью множества натуральных чисел <i>N</i> и тогда множество <i>N</i> является родителем.<br>

##### 2. Даны 4 последовательности. Необходимо:

1. Последовательность называется монотонной, если она является неубывающей или невозрастающей.
2. Если последовательность имеет предел, то она ограничена.<br>
Данные утверждения лежат в основе дальнейших исследований.<br><br>
<b>Последовательность a</b><br>
<i>а)</i>
Для исследования последовательности на монотонность вычислим разность между <i>n+1</i> и <i>n</i> членами и оценим полученный результат.
$$2^{n+1}-(n+1) - (2^n-n) = 2\times 2^n - n - 1 - 2^n +n = 2\times 2^n - 2^n - 1 = 2^n - 1$$
Разность между <i>n+1</i> и <i>n</i> членами равна: $$2^n - 1\geq1 \quad \forall n \in [1;∞),$$
то есть <i>n+1</i> будет больше <i>n</i> члена на число, которое начинается  с 1 и увеличивается по мере увеличения порядкового номера <i>n</i>. Таким образом, последовательность <i><b>а</b></i> является возрастающей и, соответственно, монотонной.<br>
<br>
<i>б)</i>
Исходя из предыдущего пункта, 
$$a_{n+1}>a_n,$$
соответственно последовательность <i>а</i> сверху не ограничена, а снизу ограничена числом 1, т.к. $$a_1=1$$

<b>Последовательность b</b><br>
<i>а)</i>
Для исследования последовательности на монотонность вычислим разность между <i>n+1</i> и <i>n</i> членами и оценим полученный результат.
$$\frac{1}{1-(n+1)}-\frac{1}{1-n}=\frac{1}{1-n-1}-\frac{1}{1-n}=-\frac{1}{n}-\frac{1}{1-n}$$
Приведем к общему знаменателю и упростим:
$$-\frac{1}{n}-\frac{1}{1-n}=\frac{-1+n-n}{n(1-n)}=\frac{-1}{n(1-n)}=\frac{-1}{n}\times \frac{1}{1-n}$$
Казалось бы, что тут простого, и я только усложнил себе задачу. Однако оценим по отдельности оба множителя:<br>
Множитель 1:
$$\frac{-1}{n}\geq \frac{-1}{2} \forall n \in [2;∞), \quad \text{а также является отрицательным числом, модуль значения которого меньше единицы и уменьшается по мере роста n}$$<br>
Множитель 2:
$$\frac{1}{1-n}\geq -1 \forall n \in [2;∞), \quad \text{также является отрицательным числом, модуль значения которого меньше единицы и уменьшается по мере роста n}$$<br>
Произведение обоих множителей дает нам положительное число.<br>
В итоге мы получили, что разница между <i>n+1</i> и <i>n</i> членами последовательности есть положительное число, которое уменьшается по мере увеличения порядкового номера <i>n</i>. Таким образом, последовательность <i><b>b</b></i> является возрастающей и, соответственно, монотонной.<br>
<br>
<i>б)</i>
Для исследования последовательности на ограниченность, вычислим её предел:
$$lim_{n=2}^∞\frac{1}{1-n}=\frac{1}{-∞}=0$$
соответственно последовательность <i>b</i> сверху ограничена числом 0, а снизу ограничена числом -1, т.к. $$b_2=-1$$
<br>
<i>в)</i>
Пятым по счету порядковым номером является <i>n=6</i>
$$b_6=\frac{1}{1-6}=-\frac{1}{5}=-0,2$$<br>
<b>Последовательность с</b><br>
<i>а)</i>
Для исследования последовательности на монотонность вычислим разность между <i>n+1</i> и <i>n</i> членами и оценим полученный результат.
$$-1^{n+1}+\sqrt{2(n+1)}-(-1^n+\sqrt{2n})=-1^{n+1}+\sqrt{2n+2}+1^n-\sqrt{2n}$$
Единица в любой степени будет равна единице, а минус не включен в скобки :) Поэтому останется разность квадратных корней<br>
$$\sqrt{2n+2}-\sqrt{2n}, \quad \text {домножим и разделим на сопряженное выражение, чтобы получить разность квадратов:}$$
$$\sqrt{2n+2}-\sqrt{2n}=\frac{(\sqrt{2n+2}-\sqrt{2n})\times(\sqrt{2n+2}+\sqrt{2n})}{\sqrt{2n+2}+\sqrt{2n}}=\frac{2n+2-2n}{\sqrt{2n+2}+\sqrt{2n}}=\frac{2}{\sqrt{2n+2}+\sqrt{2n}}>0, \quad \forall n \in [1;∞)$$<br>
В итоге мы получили, что разница между n+1 и n членами последовательности есть положительное число, которое уменьшается по мере увеличения порядкового номера <i>n</i>. Таким образом, последовательность <i><b>c</b></i> является возрастающей и, соответственно, монотонной.<br>
<i>б)</i>
Для исследования последовательности на ограниченность, вычислим её предел:
$$lim_{n=1}^∞-1^n+\sqrt{2n}=∞-1=∞$$
Таким образом сверху последовательность не ограничена, а снизу ограничена числом: $$c_1=\sqrt{2}-1$$<br>
