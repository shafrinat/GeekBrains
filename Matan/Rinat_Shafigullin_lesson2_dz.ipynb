### –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ –∫ —É—Ä–æ–∫—É 2
##### 1. –ü—Ä–æ—á–∏—Ç–∞—Ç—å –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–π –ª–æ–≥–∏–∫–∏, –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –∏—Ö –æ—Ç—Ä–∏—Ü–∞–Ω–∏—è –∏ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –∏—Å—Ç–∏–Ω–Ω–æ—Å—Ç—å 

1) $$\forall y \in [0; 1]: sgn(y) = 1$$

–î–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ —á–∏—Ç–∞–µ—Ç—Å—è –∫–∞–∫ "–î–ª—è –ª—é–±–æ–≥–æ y –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–µ–≥–æ –æ—Ç—Ä–µ–∑–∫—É –æ—Ç 0 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –¥–æ 1 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –≤–µ—Ä–Ω–æ 
—É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ, —á—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Å–∏–≥–Ω—É–º –æ—Ç —É —Ä–∞–≤–Ω–æ 1"

<b>–†–µ—à–µ–Ω–∏–µ:</b>
–§—É–Ω–∫—Ü–∏—è sgn(y) –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$$sgn(y) = \begin{cases}
                1, & \quad \text{–ø—Ä–∏ } y\text{>0,} \\
                0, & \quad \text{–ø—Ä–∏ } y\text{=0,} \\
                -1, & \quad \text{–ø—Ä–∏ } y\text{<0} 
            \end{cases}$$

–ò—Å—Ö–æ–¥—è –∏–∑ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ —Å–∏–≥–Ω—É–º, y –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å —Ä–∞–≤–µ–Ω 1 –≤ —Ç–æ—á–∫–µ 0. –ó–Ω–∞—á–∏—Ç, –¥–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ –ª–æ–∂–Ω–æ.

–ü–æ—Å—Ç—Ä–æ–∏–º –æ—Ç—Ä–∏—Ü–∞–Ω–∏–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏—è
$$\exists y \in [0; 1]: sgn(y) \neq 1$$

–î–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ —á–∏—Ç–∞–µ—Ç—Å—è –∫–∞–∫ "–°—É—â–µ—Å—Ç–≤—É–µ—Ç —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω y –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–µ–≥–æ –æ—Ç—Ä–µ–∑–∫—É –æ—Ç 0 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –¥–æ 1 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ —Ç–∞–∫–æ–π, —á—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Å–∏–≥–Ω—É–º –æ—Ç —É –Ω–µ —Ä–∞–≤–Ω–æ 1".
–ò—Å—Ö–æ–¥—è –∏–∑ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ —Å–∏–≥–Ω—É–º, –≤ —Ç–æ—á–∫–µ 0 –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞–≤–Ω–æ 0, —Ç.–µ. –Ω–µ —Ä–∞–≤–Ω–æ 1. –ó–Ω–∞—á–∏—Ç –æ—Ç—Ä–∏—Ü–∞–Ω–∏–µ –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏—è —è–≤–ª—è–µ—Ç—Å—è –∏—Å—Ç–∏–Ω–æ–π, —á—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –ª–æ–∂–Ω–æ—Å—Ç—å –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏—è.

2) $$\forall n \in {N}>2: \exists x, y, z \in {N}: x^n=y^n+z^n$$

–î–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ —á–∏—Ç–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: "–î–ª—è –ª—é–±–æ–≥–æ n, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–µ–≥–æ –º–Ω–æ–∂–µ—Å—Ç–≤—É –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª –∏ –±–æ–ª—å—à–µ 2 –≤–µ—Ä–Ω–æ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ, —á—Ç–æ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ x, y, z –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–∏—Ö –º–Ω–æ–∂–µ—Å—Ç–≤—É –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, —Ç–∞–∫–∏–µ —á—Ç–æ x –≤ —Å—Ç–µ–ø–µ–Ω–∏ n —Ä–∞–≤–Ω—è–µ—Ç—Å—è —Å—É–º–º–µ y –≤ —Å—Ç–µ–ø–µ–Ω–∏ n –∏ z –≤ —Å—Ç–µ–ø–µ–Ω–∏ n"

<b>–†–µ—à–µ–Ω–∏–µ:</b> –¢–µ–æ—Ä–µ–º–∞ –§–µ—Ä–º–∞ —É—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ –¥–ª—è –ª—é–±–æ–≥–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞ n>2 —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $$x^n=y^n+z^n$$ <b><u>–Ω–µ –∏–º–µ–µ—Ç</u></b> —Ä–µ—à–µ–Ω–∏–π –≤ —Ü–µ–ª—ã—Ö –Ω–µ–Ω—É–ª–µ–≤—ã—Ö —á–∏—Å–ª–∞—Ö x, y, z. –ó–Ω–∞—á–∏—Ç, –¥–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ –ª–æ–∂–Ω–æ.

–ü–æ—Å—Ç—Ä–æ–∏–º –æ—Ç—Ä–∏—Ü–∞–Ω–∏–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏—è:
$$\exists n \in {N}>2: \forall x, y, z \in {N}: x^n \neq y^n+z^n$$
–î–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ —á–∏—Ç–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: "–°—É—â–µ—Å—Ç–≤—É–µ—Ç —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ n, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–µ–µ –º–Ω–æ–∂–µ—Å—Ç–≤—É –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª –∏ –±–æ–ª—å—à–µ 2 —Ç–∞–∫–æ–µ, —á—Ç–æ –¥–ª—è –ª—é–±—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π x, y, z –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–∏—Ö –º–Ω–æ–∂–µ—Å—Ç–≤—É –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, –≤–µ—Ä–Ω–æ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ, —á—Ç–æ x –≤ —Å—Ç–µ–ø–µ–Ω–∏ n –Ω–µ —Ä–∞–≤–Ω—è–µ—Ç—Å—è —Å—É–º–º–µ y –≤ —Å—Ç–µ–ø–µ–Ω–∏ n –∏ z –≤ —Å—Ç–µ–ø–µ–Ω–∏ n"

–í—ã–±–µ—Ä–µ–º –¥–ª—è x –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ–µ 1, –∞ y –∏ z –Ω–µ —Ä–∞–≤–Ω—ã–µ 1 –ø–æ–ª—É—á–∏–º, —á—Ç–æ 1 –≤ —Å—Ç–µ–ø–µ–Ω–∏ n –Ω–µ —Ä–∞–≤–Ω–æ —Å—É–º–º–µ –ª—é–±—ã—Ö –¥—Ä—É–≥–∏—Ö –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª –≤ —Å—Ç–µ–ø–µ–Ω–∏ n, —á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è –∏—Å—Ç–∏–Ω–æ–π, –∏ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –ª–æ–∂–Ω–æ—Å—Ç—å –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏—è.

3) $$\forall x \in {R} \exists X \in {R}: X>x$$

–î–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ —á–∏—Ç–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: "–î–ª—è –ª—é–±–æ–≥–æ <i>x</i>, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–µ–≥–æ –º–Ω–æ–∂–µ—Å—Ç–≤—É –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ <i>X</i>, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–∏–π –º–Ω–æ–∂–µ—Å—Ç–≤—É –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª, —Ç–∞–∫–æ–µ, —á—Ç–æ <i>X</i> –±–æ–ª—å—à–µ <i>x</i>".

<b>–†–µ—à–µ–Ω–∏–µ:</b> –û–¥–Ω–∏–º –∏–∑ —Å–≤–æ–π—Å—Ç–≤ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª —è–≤–ª—è–µ—Ç—Å—è –µ–≥–æ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç—å. –≠—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ –∫ –ª—é–±–æ–º—É —ç–ª–µ–º–µ–Ω—Ç—É –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –≤—Å–µ–≥–¥–∞ –º–æ–∂–Ω–æ –ø—Ä–∏–±–∞–≤–∏—Ç—å –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –ø–æ–ª—É—á–∏—Ç—å –Ω–æ–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç –º–Ω–æ–∂–µ—Å—Ç–≤–∞, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç –±–æ–ª—å—à–µ –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ. –û—Ç—Å—é–¥–∞ —Å–ª–µ–¥—É–µ—Ç, —á—Ç–æ –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ –∏—Å—Ç–∏–Ω–Ω–æ.

–ü–æ—Å—Ç—Ä–æ–∏–º –æ—Ç—Ä–∏—Ü–∞–Ω–∏–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏—è:
$$\exists x \in {R} \forall X \in {R}: X\leq x$$

–î–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ —á–∏—Ç–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: "–°—É—â–µ—Å—Ç–≤—É–µ—Ç —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ <i>x</i>, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–µ–µ –º–Ω–æ–∂–µ—Å—Ç–≤—É –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª, –ø—Ä–∏ —ç—Ç–æ–º –¥–ª—è –ª—é–±—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π <i>X</i>, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–µ–≥–æ –º–Ω–æ–∂–µ—Å—Ç–≤—É –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª, –±—É–¥–µ—Ç —Å–ø—Ä–∞–≤–µ–¥–ª–∏–≤–æ, —á—Ç–æ <i>X</i> –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ <i>x</i>".
–ò–Ω—ã–º–∏ —Å–ª–æ–≤–∞–º–∏, –∫–∞–∫–æ–µ –±—ã –∑–Ω–∞—á–µ–Ω–∏–µ <i>X</i> –∏–∑ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª –º—ã –Ω–∏ –≤–∑—è–ª–∏, –æ–Ω–æ –≤—Å–µ–≥–¥–∞ –±—É–¥–µ—Ç –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ –∑–Ω–∞—á–µ–Ω–∏—è <i>x</i> –∏–∑ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª. –ü—Ä–æ—â–µ –≥–æ–≤–æ—Ä—è, —É –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª –µ—Å—Ç—å –∫–æ–Ω–µ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –æ–Ω–æ —Ä–∞–≤–Ω–æ <i>x</i>, –Ω–æ —ç—Ç–æ –ø—Ä–æ—Ç–∏–≤–æ—Ä–µ—á–∏—Ç —Å–≤–æ–π—Å—Ç–≤—É –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –æ—Ç—Ä–∏—Ü–∞–Ω–∏–µ –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏—è –ª–æ–∂–Ω–æ, –∞ —ç—Ç–æ –¥–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∏—Å—Ç–∏–Ω–Ω–æ—Å—Ç—å –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏—è.

4) $$\forall x \in {C} \nexists y \in {C}: x>y\|x<y$$

–î–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ —á–∏—Ç–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: "–î–ª—è –ª—é–±–æ–≥–æ x, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–µ–≥–æ –º–Ω–æ–∂–µ—Å—Ç–≤—É –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–µ–ª –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç y, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–µ–≥–æ –º–Ω–æ–∂–µ—Å—Ç–≤—É –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–µ–ª, —Ç–∞–∫–æ–µ —á—Ç–æ x –±–æ–ª—å—à–µ y –∏–ª–∏ x –º–µ–Ω—å—à–µ y". –ò–Ω—ã–º–∏ —Å–ª–æ–≤–∞–º–∏, –¥–∞–Ω–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ —É—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ –≤–æ –º–Ω–æ–∂–µ—Å—Ç–≤–µ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–µ–ª –ª—é–±–æ–µ —á–∏—Å–ª–æ x —Ä–∞–≤–Ω–æ –ª—é–±–æ–º—É —á–∏—Å–ª—É y. 

<b>–†–µ—à–µ–Ω–∏–µ:</b> –ú–Ω–æ–∂–µ—Å—Ç–≤–æ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–µ–ª –≤–∫–ª—é—á–∞–µ—Ç –≤ —Å–µ–±—è —Ç–∞–∫–∂–µ –¥—Ä—É–≥–∏–µ –º–Ω–æ–∂–µ—Å—Ç–≤–∞, –≤ —Ç–æ–º —á–∏—Å–ª–µ –∏ –º–Ω–æ–∂–µ—Å—Ç–≤–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª N, —ç–ª–µ–º–µ–Ω—Ç—ã –∫–æ—Ç–æ—Ä–æ–≥–æ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ –∏ –ø—Ä–∏–º–µ–Ω—è—é—Ç—Å—è –¥–ª—è —Å—á–µ—Ç–∞ –ø—Ä–µ–¥–º–µ—Ç–æ–≤:
$${N} = \{1, 2, 3, 4, ... n, n+1 \}$$
–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–µ –≤—ã—Å–∫–∞–∑—ã–≤–∞–Ω–∏–µ –ª–æ–∂–Ω–æ.

$$B \backslash C = \{2, '—à–µ—Å—Ç—å'\}$$
$$C \backslash B = \{9, '–¥–µ—Å—è—Ç—å'\}$$
$$A \backslash C = \{0, 1, 2\}$$
$$C \backslash A = \{5, 9, '—Å–µ–º—å', '–¥–µ—Å—è—Ç—å'\}$$

<b>–°–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å</b>
$$A \Delta B = \{0, 1, 5, '—à–µ—Å—Ç—å', '—Å–µ–º—å'\}$$
$$A \Delta C = \{0, 1, 2, 5, 9, '—Å–µ–º—å', '–¥–µ—Å—è—Ç—å'\}$$
$$B \Delta C = \{2, 9, '—à–µ—Å—Ç—å', '–¥–µ—Å—è—Ç—å'\}$$

```python
# It is Python time, let the coding begin

# –ó–∞–¥–∞–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –ê, –í, –°
A = {0, 1, 2, 3, 4}
B = {2, 3, 4, 5, '—à–µ—Å—Ç—å', '—Å–µ–º—å'}
C = {3, 4, 5, 9, '—Å–µ–º—å', '–¥–µ—Å—è—Ç—å'}

#–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ (–¥–∏–∑—ä—é–Ω–∫—Ü–∏—è)
print(f'–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ (–¥–∏–∑—ä—é–Ω–∫—Ü–∏—è) –º–Ω–æ–∂–µ—Å—Ç–≤ –ê –∏ –í = {B.union(A)}')
print(f'–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ (–¥–∏–∑—ä—é–Ω–∫—Ü–∏—è) –º–Ω–æ–∂–µ—Å—Ç–≤ –ê –∏ C = {C.union(A)}')
print(f'–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ (–¥–∏–∑—ä—é–Ω–∫—Ü–∏—è) –º–Ω–æ–∂–µ—Å—Ç–≤ B –∏ C = {C.union(B)}')
print('') #–ü—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞
#–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ (–∫–æ–Ω—ä—é–Ω–∫—Ü–∏—è)
print(f'–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ (–∫–æ–Ω—ä—é–Ω–∫—Ü–∏—è) –º–Ω–æ–∂–µ—Å—Ç–≤ –ê –∏ –í = {B.intersection(A)}')
print(f'–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ (–∫–æ–Ω—ä—é–Ω–∫—Ü–∏—è) –º–Ω–æ–∂–µ—Å—Ç–≤ –ê –∏ C = {C.intersection(A)}')
print(f'–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ (–∫–æ–Ω—ä—é–Ω–∫—Ü–∏—è) –º–Ω–æ–∂–µ—Å—Ç–≤ B –∏ C = {C.intersection(B)}')
print(f'–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ (–∫–æ–Ω—ä—é–Ω–∫—Ü–∏—è) –º–Ω–æ–∂–µ—Å—Ç–≤ –ê, B –∏ C = {C.intersection(B.intersection(A))}')
print('') #–ü—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞
#–†–∞–∑–Ω–æ—Å—Ç—å
print(f'–†–∞–∑–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤ –ê –∏ –í = {A.difference(B)}')
print(f'–†–∞–∑–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤ B –∏ A = {B.difference(A)}')
print(f'–†–∞–∑–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤ B –∏ C = {B.difference(C)}')
print(f'–†–∞–∑–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤ C –∏ B = {C.difference(B)}')
print(f'–†–∞–∑–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤ –ê –∏ C = {A.difference(C)}')
print(f'–†–∞–∑–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤ C –∏ A = {C.difference(A)}')
print('') #–ü—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞
#–°–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å
print(f'–°–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤ –ê –∏ –í = {B.symmetric_difference(A)}')
print(f'–°–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤ –ê –∏ C = {C.symmetric_difference(A)}')
print(f'–°–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤ B –∏ C = {C.symmetric_difference(B)}')
```
